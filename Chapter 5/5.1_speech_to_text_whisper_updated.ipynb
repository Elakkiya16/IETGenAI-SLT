# 5.1 Speech to Text using Whisper

This notebook demonstrates how to transcribe a speech audio file into text using the Whisper model. It also includes audio format conversion for compatibility.

```python
# ✅ Step 3: Import libraries
import whisper
import torchaudio
import librosa
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
from pydub import AudioSegment
```

```python
# ✅ Step 4: Convert .mp3 to .wav
mp3_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/could _you_please_talk_slower.mp3'
wav_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/could_you_please_talk_slower.wav'

audio = AudioSegment.from_mp3(mp3_path)
audio.export(wav_path, format='wav')

print(f"Converted and saved: {wav_path}")
```

```python
# ✅ Step 5: Play converted WAV audio
ipd.Audio(wav_path)
```

```python
# ✅ Step 6: Plot waveform
y, sr = librosa.load(wav_path)
plt.figure(figsize=(10, 3))
plt.plot(y)
plt.title('Waveform of Converted Audio')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# ✅ Step 7: Load Whisper and transcribe
model = whisper.load_model('small')
result = model.transcribe(wav_path)
print("Transcription:", result['text'])
```

```python
# ✅ Step 8: Save the transcription to a file
transcription_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/transcription.txt'

with open(transcription_path, 'w') as f:
    f.write(result['text'])

print(f"Transcription saved to: {transcription_path}")
```

```python
# ✅ Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ✅ Step 2: Install required packages
!pip install -q git+https://github.com/openai/whisper.git
!pip install -q librosa torchaudio pydub

# ✅ Step 3: Import libraries
import whisper
import torchaudio
import librosa
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
from pydub import AudioSegment

# ✅ Step 4: Convert .mp3 to .wav
mp3_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/could _you_please_talk_slower.mp3'
wav_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/could_you_please_talk_slower.wav'

audio = AudioSegment.from_mp3(mp3_path)
audio.export(wav_path, format='wav')

print(f"Converted and saved: {wav_path}")

# ✅ Step 5: Play converted WAV audio
ipd.Audio(wav_path)

# ✅ Step 6: Plot waveform
y, sr = librosa.load(wav_path)
plt.figure(figsize=(10, 3))
plt.plot(y)
plt.title('Waveform of Converted Audio')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# ✅ Step 7: Load Whisper and transcribe
model = whisper.load_model('small')
result = model.transcribe(wav_path)
print("Transcription:", result['text'])

# ✅ Step 8: Save the transcription to a file
transcription_path = '/content/drive/MyDrive/IETGenAI-SLT/Chapter 5/transcription.txt'

with open(transcription_path, 'w') as f:
    f.write(result['text'])

print(f"Transcription saved to: {transcription_path}")
```
